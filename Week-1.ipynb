### Install a Python package for Networkx

```python
import networkx
```

```python
G = networkx.Graph()
```

### Add node to Graph -- > add_node

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)
```

### Display all Nodes inside a Graph --> nodes

```python
G.nodes()
```

### Add Edges between nodes --> add_edges

```python
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(4,6)
G.add_edge(5,4)
G.add_edge(2,3)
G.add_edge(2,6)
```

### Display all edges inside a Graph --> edges

```python
G.edges()
```

### Display a Graph

```python
import matplotlib.pyplot as plt
import networkx as nx
nx.draw(G)
plt.show()
```

```python
nx.draw(G,with_labels=1)
```

### Drawing a Complete Graph

```python
comp_Graph = nx.complete_graph(10)
nx.draw(comp_Graph,with_labels=1)
plt.show()
```

**Number of Edges**  --> .size()  
**Number of Nodes**  --> order()

```python
print("Number of Edges",comp_Graph.size())
```

```python
print("Number of Nodes",comp_Graph.order())
```

```python
H = nx.complete_graph(100)
nx.draw(H,with_labels=1)
```

Generate a Gaph with edges probability 0.5

```python
G=nx.gnp_random_graph(20,0.5)
nx.draw(G,with_labels=1)
plt.show()
print("Edges count ",G.size())
print("Node count ",G.order())
```
